```python
# Importanto bibliotecas
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')

import matplotlib.pyplot as plt
```

```python
# Importando base de dados
link = "https://raw.githubusercontent.com/gfsilveira/ebac/main/"
link += "03-Cientista%20de%20Dados/02-Crisp-DM/M%C3%B3dulo-06/b1abb33c4f.csv"
df = pd.read_csv(link)
df.head()
```

```python
# Verirficando informações
df.info()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Box plot da idade da mãe
plt.figure(figsize=(5,10))

plt.title("Idade da Mãe em anos no momento do parto")

sns.boxplot(y='IDADEMAE', data=df)

plt.ylabel("Idade da Mãe (anos)")

plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Boxplot do peso do babê
plt.figure(figsize=(5,10))

plt.title("Peso do Bebê ao nascer")

sns.boxplot(y='PESO', data=df)

plt.ylabel("Peso do Bebê (gramas)")

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Histograma da idade da mãe
f = sns.displot(x='IDADEMAE', aspect=4, kde=True, bins=20, data=df)

f.set(
    title = "Idade da Mãe em anos no momento do parto",
    xlabel = "Idade da Mãe (anos)",
    ylabel = "Número de mães por faixa de idade"
)
```

#### 4) Faça um histograma do peso do bebê

```python
# Histograma do peso do bebê
f = sns.displot(x='PESO', aspect=4, kde=True, bins=20, data=df)

f.set(
    title = "Peso do Bebê ao nascer",
    xlabel = "Peso do Bebê (gramas)",
    ylabel = "Número por faixa de peso"
)
```
