### Suposies do modelo

A regresso, vista como um modelo estatstico,  um modelo paramtrico, ou seja, nele se supe uma distribuio de probabilidades especfica para os dados.  verdade que esta distribuio , normalmente, a mais razovel para variveis contnuas, e dependendo do que queremos fazer, o teorema central do limite nos protege para muitos usos.

Nosso objetivo nesta aula  refletir sobre essas suposies, as implicaes de no cumprimento dessas suposies.

**Estrutura**

O modelo de regresso mltipla clssico tem essa forma:

$$y_i \sim \alpha + \beta_1 x_{i,1}+ \beta_2 x_{i,2}+...+ \beta_p x_{i,p} + \epsilon_i $$

com:

- $y_i$  a varivel resposta, endgena ou dependente.  ela que vamos explicar/prever com base nas demais variveis
- $x_{i,j}$ so as variveis explicativas, exgenas ou dependentes ( tudo sinimo). Cada $x_{i,j}$  o valor da varivel $j$ para o indivduo $i$ com $i=1, 2,...n$ e $j=1, 2, ... p$.
- $\alpha$  uma constante, que representa o valor mdio de $y$ quando todos os $x$ so iguais a zero.
- $\beta_j$ so constantes que podem ser interpretadas como o aumento em $y$ para cada unidade do respectivo $x_j$ que aumentamos.
- $\epsilon$  um erro aleatrio dado pela diferena entre o valor $y_i$ e o seu respectivo valor mdio $E(y_i)$ dado pelo modelo de regresso.

As suposies estatsticas dessa regresso so:

- Independncia: $\epsilon_i$ so independentes entre s
- Distribuio Normal: $\epsilon_i \sim N(0,\sigma_m)$ com i=1, 2,... n e $\sigma_m$  uma constante
- Homocedasticidade: $\epsilon_i$ possuem todos a mesma varincia $\sigma_m^2$

Vamos comear por essas trs suposies e ponderar sobre os impactos de no cumprimento.

#### Independncia

A forma mais prtica de verificar a independncia  pelo grfico de resduos. Os resduos no devem apresentar qualquer tipo de padro com relao a qualquer outra varivel. As checagens mais comuns so atravs da observao de grficos de resduos versus:

- A ordem dos dados
- Valores preditos
- Variveis explicativas
- Qualquer outra varivel

Fontes comuns de dependncia podem ser encontradas na estrutura de coleta dos dados, por exemplo se h vrias medidas de um mesmo indivduo, se h grupos familiares ou outros fatores que possam relacionar as observaes. Se os resduos para o mesmo indivduo apresentarem algum tipo de padro, significa que a independncia est sendo violada.

````python
sns.residplot(x=reg.predict(), y='tip', data=df, lowess=True, 
                     scatter_kws={'alpha': 0.5}, 
                     line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8});
````

#### Distribuio Normal

**Pra que  importante?**<br>
Essa suposio  importante para garantir que os estimadores tenham as distribuies estudadas. Com fugas da distribuio normal podem no valer mais as inferncias sobre os parmetros e sobre a varivel resposta para novas observaes.

**Impactos**<br>
Primeiramente, observo que no  um grande pecado violar a normalidade quando o nosso objetivo  fazer previses. Se os resduos no tiverem distribuio normal vo estar comprometidas as inferncias sobre o *erro padro* das previses (principalmente), e tambm sobre os erros padro dos parmetros. Lembrando que estes ltimos so mais protegidos pelo teorema central do limite, de modo que para amostras grandes haver uma interferncia um pouco menor.

**Verificao**
A checagem dessa distribuio  feita basicamente em se avaliando os resduos. O resduo *ordinrio*  simplesmente $y_i-\hat{y}_i$, mas h uma variedade de outras proposies de resduos tambm.

Vamos falar tambm do *studentized residual*, que  o resduo dividido pelo seu erro padro. Se todas as suposies da regresso esto atendidas, ele tem distribuio normal padro, o que torna conveniente de se avaliar a sua amplitude - por exemplo, esperamos que a maioria dos resduos estejam entre -2 e 2, e muito poucos mesmo estejam fora do intervalo \[-3, 3\].

**Medidas corretivas** <br>
Um caso comum de fuga dessa suposio  quando temos assimetria nos dados. Uma medida corretiva popular  transformar a varivel resposta com uma funo ```log```, o que corrige em grande parte dos casos a distribuio. Alm disso, h opes de modelos mais robustos que ainda permitem os testes, mas no vamos abord-los neste mdulo.

### Coeficiente v de Cramer

Vulgarmente chamado de "correlao de Cramer",  mais precisamente uma medida de associao. Mede a associao entre variveis qualitativas, baseado na estatstica *qui-quadrado*.

- **Qui-quadrado**: A medida mais bsica de associao entre variveis qualitativas  a chamada "estatstica qui-quadrado", construda a partir de uma tabela cruzada entre as duas variveis. Sob $H_0$ (as variveis so independentes), essa medida tem uma distribuio qui-quadrado. O problema  que essa medida no  muito intuitiva por variar de 0 a $+\infty$ e no  comparvel entre tabelas de tamanhos diferentes. 

- **V de Cramer**: Baseada na estatstica qui-quadrado, ela busca uma medida que seja comparvel para qualquer tamanho de tabela cruzada e  mais intuitiva de se ler: varia entre 0 e 1, quanto mais prxima de 1, maior a associao e vice-versa.

O *v de Cramer* no est implementado em Python, mas abaixo h uma funo que foi publicada no *Journal of the Korean Statistical Society*. 

````python
def cramers_corrected_stat(x,y):
    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    result=-1
    if len(x.value_counts())==1 :
        print("First variable is constant")
    elif len(y.value_counts())==1:
        print("Second variable is constant")
    else:   
        conf_matrix=pd.crosstab(x, y)

        if conf_matrix.shape[0]==2:
            correct=False
        else:
            correct=True

        chi2 = ss.chi2_contingency(conf_matrix, correction=correct)[0]

        n = sum(conf_matrix.sum())
        phi2 = chi2/n
        r,k = conf_matrix.shape
        phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
        rcorr = r - ((r-1)**2)/(n-1)
        kcorr = k - ((k-1)**2)/(n-1)
        result=np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
        return result
````

Esta  uma adaptao da funo acima para funcionar para variveis contnuas tambm. Como ela s funciona para variveis qualitativas, ela categoriza a varivel contnua para calcular a medida.

````python
def cram(x, y, max_cat=5):
    if (np.dtype(x) in {np.dtype('float64'), np.dtype('int64')}) \
        and (len(np.unique(x)) > max_cat):
            x_ = pd.qcut(x, max_cat)
#             print('#########')
    else: x_ = x
    if (np.dtype(y) in {np.dtype('float64'), np.dtype('int64')}) \
        and (len(np.unique(y)) > max_cat):
            y_ = pd.qcut(y, max_cat)
    else: y_ = y
    return cramers_corrected_stat(x_ ,y_)
````

### Variance Inflation Factor

O *Variance Inflation Factor* (ou VIF)  uma das medidas de multicolinearidade mais populares. Essa medida indica o quanto da variabilidade de uma varivel  explicada pelas demais. Se as variveis "se explicam" muito umas s outras, vamos ter multicolinearidade.

O VIF indica quanto de uma varivel X(i)  explicada pelas demais, ele  construdo assim:
- Faz-se um modelo de regresso para explicar X(i) (como resposta) com todas as demais X1, X2, etc (exceto X(i)) como explicativas
- Calculamos o $R^2$
- O VIF  $1/R^2$

Assim, um VIF igual a 20 significa que 95% da variabilidade  explicada pelas demais variveis. VIF=10 corresponde a 90% de explicao da varincia. Dessa forma ele serve de alerta.

**Medidas corretivas**<br>

A medida corretiva mais comum  utilizar apenas uma das variveis no modelo. Podemos utilizar tambm a PCA, ou extrair a informao exclusiva da varivel como veremos mais adiante.

### Construindo variveis ortogonais

Uma forma de se corrigir a multicolinearidade  a construo de variveis ortogonais. Podemos atingir isso usando anlise de componentes principais (PCA), ou tentar remover a varincia de uma varivel explicada por outra (ou outras).

Como fazemos isso? Vamos tomar a varivel ```bill_length_mm``` por exemplo. Para cada espcie, vamos subtrair do valor dessa varivel, a mdia dela dentro da espcie. Parece complicada a frase, mas o procedimento  simples: Vamos calcular a mdia de ```bill_length_mm``` para todos os pinguns *Gentoo*, e fazer ```bill_length_mm``` menos o valor encontrado - e o mesmo para as demais espcies.

A interpretao da varivel passa a ser "Quantos centmetros maior (ou menor) que a mdia da sua espcie este pinguin tem?". Isso faz com que essa varivel no seja mais explicada pela espcie, o que deve reduzir bem a multicolinearidade.

````python
penguins.groupby('species')[['bill_length_mm','bill_depth_mm','flipper_length_mm']].mean()
````

````python
medias_bl = penguins['bill_length_mm'].groupby(penguins.species).mean()
penguins['bl_sp'] = penguins['bill_length_mm'] - penguins['species'].map(medias_bl)

medias_bd = penguins['bill_depth_mm'].groupby(penguins.species).mean()
penguins['bd_sp'] = penguins['bill_depth_mm'] - penguins['species'].map(medias_bd)

medias_bf = penguins['flipper_length_mm'].groupby(penguins.species).mean()
penguins['fl_sp'] = penguins['flipper_length_mm'] - penguins['species'].map(medias_bf)

penguins[['bl_sp', 'bd_sp', 'fl_sp']].corr(method='spearman')
````

````python
sns.pairplot(penguins[['bl_sp', 'bd_sp', 'fl_sp', 'species']], hue='species');
````

````python
r1 = smf.ols('''
    body_mass_g ~ 
        species + sex 
        + bl_sp
        + bd_sp
        + fl_sp
    ''', data=penguins).fit()
r1.summary()
````
