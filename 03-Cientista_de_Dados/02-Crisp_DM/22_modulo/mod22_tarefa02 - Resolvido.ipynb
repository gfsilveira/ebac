3. Monte um gráfico de linhas para as médias móveis dos ativos da sua base

````python
df_adj_load_mean90 = df_ativo_load['Adj Close']
df_adj_load_mean90_vertical = df_adj_load_mean90.stack().reset_index()
df_adj_load_mean90_vertical.columns = ['data', 'ativo', 'valor']
fig = px.line(
    df_adj_load_mean90_vertical,
    x="data",
    y="valor",
    color='ativo',
)
fig.update_layout(
    title="Adjusted Close",
    xaxis_title="Datas",
    yaxis_title="Valores",
    legend=dict(title="Ativos"),
)
fig.show()
````

````python
df_adj_load_mean90 = df_ativo_load['Adj Close'].rolling(90).mean()
df_adj_load_mean90_vertical = df_adj_load_mean90.stack().reset_index()
df_adj_load_mean90_vertical.columns = ['data', 'ativo', 'valor']
fig = px.line(
    df_adj_load_mean90_vertical,
    x="data",
    y="valor",
    color='ativo'
)
fig.update_layout(
    title="Adjusted Close, moving average (90 days)",
    xaxis_title="Datas",
    yaxis_title="Valores",
    legend=dict(title="Ativos"),
)
fig.show()
````

4. Monte um gráfico de linhas para os desvios padrão em janela móvel

````python
df_adj_load_std90 = df_ativo_load['Adj Close'].rolling(90).std()
df_adj_load_std90_vertical = df_adj_load_std90.stack().reset_index()
df_adj_load_std90_vertical.columns = ['data', 'ativo', 'valor']
fig = px.line(
    df_adj_load_std90_vertical,
    x="data",
    y="valor",
    color='ativo'
)
fig.update_layout(
    title="Adjusted Close, moving standard deviation (90 days)",
    xaxis_title="Datas",
    yaxis_title="Valores",
    legend=dict(title="Ativos"),
)
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Lista de dindicadores sem repetição e de ativo
ticker = 0
lista_indicadores = list({i[0] for i in df_ativo_load.columns})
print(lista_indicadores)
ativo = [df_ativo_load.columns[ticker][1]]*len(lista_indicadores)
print(ativo)
````

````python
# Agrupando as informações
colunas = list(zip(lista_indicadores, ativo))
colunas
````

````python
# Obtendo o mesmo resultado anterior com fatiamento
colunas = list(df_ativo_load.columns[ticker::3])
colunas
````

````python
# Selecionando os dados dos últimos 60 dias
bvsp = df_ativo_load.loc['2022-10-30':, colunas]
bvsp.columns = lista_indicadores
bvsp.head()
````

````python
# Determinando os hiperpaâmetros para o candlestick
graph = {
    'x': bvsp.index,
    'open': bvsp.Open,
    'close': bvsp.Close,
    'high': bvsp.High,
    'low': bvsp.Low,
    'type': 'candlestick',
    'name': df_ativo_load.columns[ticker][1],
    'showlegend': True
}
````

````python
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Índice NASDAq",
    # layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
