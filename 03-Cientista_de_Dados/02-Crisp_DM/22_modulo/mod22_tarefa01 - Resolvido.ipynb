# 2 Observando a volatilidade dentro do dia

```python
# Obtendo os dados do disco
df_ativo_load = load("./data/df_ativo")
df_ativo_load.head()
```

## 2.1 Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).

```python
# Verificando colunas
df_ativo_load.columns
```

```python
# Calculando a amplitude
colunas_spred = [
    ('Spread','IVV'),
    ('Spread','QQQ'),
    ('Spread','SCHD')
]
df_ativo_load[colunas_spred] = df_ativo_load['High'] - df_ativo_load['Low']
df_ativo_load['Spread'].head()
```

## 2.2 Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.

```python
# Organizando a DataFrame para o plot
spread = df_ativo_load['Spread'].stack().reset_index()
spread.columns = ['Data', 'Ativo', 'Valor']

spread.head()
```

```python
# Plotando o gráfico de amplitude
fig = px.line(spread, x="Data", y="Valor", color='Ativo', template='plotly_white')
fig.show()
```

## 2.3 Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Verificando a amplitude nas amplitudes de IVV
coluna = "IVV"
max = spread[spread['Ativo'] == coluna]['Valor'].max()
min = spread[spread['Ativo'] == coluna]['Valor'].min()
max - min
```

```python
# Verificando a amplitude nas amplitudes de QQQ
coluna = "QQQ"
max = spread[spread['Ativo'] == coluna]['Valor'].max()
min = spread[spread['Ativo'] == coluna]['Valor'].min()
max - min
```

```python
# Verificando a amplitude nas amplitudes de SCHD
coluna = "SCHD"
max = spread[spread['Ativo'] == coluna]['Valor'].max()
min = spread[spread['Ativo'] == coluna]['Valor'].min()
max - min
```

### Conclusão
- O ativo com a maior valatilidade no período foi o QQQ, com spread de 21.37
