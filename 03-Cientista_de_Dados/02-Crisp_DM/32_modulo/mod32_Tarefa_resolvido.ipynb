```python
import numpy as np
from scipy import stats

import math

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

from scipy.stats import norm, t
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

$$ Z = \frac{X - \mu}{\sigma} $$

1. Parmetros
   - $\mu$ = 15kg mdia da distribuio
   - $\sigma$ = 0,1kg desvio padro
   - $X$ > 15,2kg = probabilidade de $X$ exceder 15,2kg

2. Padronizar o valor $X$ = 15,2
$$
Z = \frac{X - \mu}{\sigma} = \frac{15,2 - 15}{0.1} = \frac{0.2}{0.1} = 2
$$

3. Probabilidade correspondente em $Z$<br>
$Z$ = 2 tm probabilidade acumulada 0,9772<br>
$Z$ > 2,  1 - 0,9772 = 0,0228

4. Conclusso<br>
A probabilidade de que um pacote tirado aleatoriamente pese mais que 15,2 kg  <b>2,28%</b>

```python
mu = 15 # mdia
sigma = 0.1 # desvio padro
x = 15.2 # valor limite

# Calcula a probabilidade P(X > x)
prob = 1 - norm.cdf(x, loc=mu, scale=sigma)
print(f"A probabilidade de que um pacote tirado aleatoriamente pese mais que {x} kg  {round(prob*100, 2)}%.")
```

```python
# Configurao dos eixos
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

# Intervalo para os grficos
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)

# ** Grfico da PDF (funo densidade de probabilidade) **
y = norm.pdf(x, mu, sigma) # Valores da PDF
low = 15.2

# Plota a PDF
ax[0].plot(x, y, label="Densidade de Probabilidade (PDF)")
# Sombreia a rea da probabilidade
ax[0].fill_between(
    x,
    y,
    where=(low < x),
    alpha=0.5,
    color='blue',
    label="rea da Probabilidade"
)
ax[0].set_title("Funo Densidade de Probabilidade (PDF)")
ax[0].set_xlabel("Peso (kg)")
ax[0].set_ylabel("Probabilidade")
ax[0].legend()

# ** Grfico da CDF (funo de distribuio acumulada) **
cdf = norm.cdf(x, mu, sigma) # Valores da CDF
ax[1].plot(x, cdf, label="Distribuio Acumulada (CDF)")

# Marca os limites no grfico da CDF
ax[1].vlines(
    x=[low],
    ymin=0,
    ymax=norm.cdf(low, mu, sigma),
    colors='r',
    linestyles='dotted',
    label="Limite"
)
ax[1].hlines(
    y=norm.cdf(low, mu, sigma),
    xmin=mu - 3 * sigma,
    xmax=low,
    colors='r',
    linestyles='dotted'
)

# Formatao de percentual no eixo Y
ax[1].yaxis.set_major_formatter(mtick.PercentFormatter(1))
ax[1].set_title("Funo de Distribuio Acumulada (CDF)")
ax[1].set_xlabel("Peso (kg)")
ax[1].set_ylabel("Probabilidade Acumulada")
ax[1].legend()

# Calcula a probabilidade para a rea destacada
prob_ar = 1 - norm.cdf(low, mu, sigma)
m = f'A probabilidade para valores maiores que {low:.2f} kg  de {prob_ar:.2%}'
print(m.replace('.', ','))

# Exibe os grficos
plt.tight_layout()
plt.show()
```
