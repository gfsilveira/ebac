## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

### 1.1 (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.

````python
import pandas as pd
import numpy as np
````

````python
# Obtendo o número de acertos
np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

# Nome das colunas
notas_columns = [
    "df_mat",
    "df_por",
    "df_geo",
    "df_ing",
    "df_his",
    "df_fis",
    "df_qui",
]

# Instanciando uma DataFrame com todos os acertos
df_notas = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)
df_notas.columns = notas_columns
df_notas.head()
````

````python
# 1) Seu código aqui

# Instanciando com o número de questões de cada matéria
assunto	= [
    "Matemática",
    "Português",
    "Geografia",
    "Inglês",
    "História",
    "Física",
    "Química",
]

numero_questoes = [
    24,
    18,
    8,
    8,
    8,
    12,
    12,
]

codigo	= [
    "df_mat",
    "df_por",
    "df_geo",
    "df_ing",
    "df_his",
    "df_fis",
    "df_qui",
]
numero_questoes = pd.DataFrame(columns=["assunto","número de questões","código"], data=zip(assunto, numero_questoes, codigo))
numero_questoes
````

````python
# Calcula a porcentagem para cada uma das colunas da DataFrame com os acertos.
coluna = 0
while coluna < len(numero_questoes):

  # Obtem o nome da nova coluna, com base no nome da coluna calculada
  nome_coluna = numero_questoes.iloc[coluna,2] + "_porc"

  # Obtem a porcentagem, dividindo o número de acertos pelo total de questões, multiplicando por 100.
  valor = round((df_notas.iloc[:,coluna] / numero_questoes.iloc[coluna,1])*100, 2)

  # Insere as porcentagens na DataFrame com as notas
  df_notas[nome_coluna] = valor

  # Soma para ser utilizada como index
  coluna += 1
df_notas.head()
````

````python
# Para organizar a DataFrame com as notas, no formato acerto, porcentagem

# Obtem uma lista com a primeira metade ds primeiros index de colunas
freq_abso = [i for i in range(int(len(df_notas.columns)/2))]

# Obtem uma lista com a segunda metade ds primeiros index de colunas
freq_rela = [j for j in range(int(len(df_notas.columns)/2), len(df_notas.columns))]

# Concatena as listas
n_colunas = freq_abso + freq_rela

# Intercala as listas
n_colunas[::2] = freq_abso
n_colunas[1::2] = freq_rela

n_colunas
````

````python
# Usa a lista com os index intercalados para reorganizar a DataFrame
df_notas = df_notas[df_notas.columns[n_colunas]]
df_notas.head()
````

### 1.2 (operações entre *DataFrames) Calcule o total de acertos de cada aluno.

````python
# 2) Seu código aqui

# O total de acertos é somado para cada aluno, inserindo o dado na DataFrame
df_notas['soma_acertos'] = df_notas.select_dtypes("int64").T.sum()
df_notas.head()
````

### 1.3 Calcule o porcentual geral de cada aluno.

````python
# 3) Seu código aqui

# O total de acertos (última coluna) é dividido pelo totral de questões, e inserido no final da data
df_notas['porc_geral'] = round((df_notas['soma_acertos'] / numero_questoes['número de questões'].sum())*100, 2)
df_notas.head()
````

### 1.4 Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?

````python
# 4) Seu código aqui

# Filtra apenas alunos com porcentagem de acertos superior ou igual a 45
print(f"{(df_notas['porc_geral'] >= 45).sum()} alunos foram aprovados")
````

### 2.1 Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.

````python
# seu código aqui
# São obtidos os alunos com nehhum acerto em física ou matemática ou química
df_notas_desclass = df_notas[
                      (df_notas['df_mat'] == 0) |
                      (df_notas['df_fis'] == 0) |
                      (df_notas['df_qui'] == 0)
]
df_notas_desclass.head()
````

### 2.2 Quantos são esses alunos?

````python
# Obtem o número de alunos da DataFrame de desclassificados
print(f"São {len(df_notas_desclass)} alunos com nehhum acerto em física ou matemática ou química!")
````

### 2.3 Qual a média desses alunos em história e geografia?

````python
# Obtem a média das notas de hsitória e geografia de todos os alunos desclassificados
media_his, media_geo = df_notas_desclass[['df_his_porc',"df_geo_porc"]].mean().values.round(2)
print(f"Esses alunos têm média de {media_his}% em história e {media_geo}% em geografia")
````

### 2.4 Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# Seleciona todos os alunos que não zeraram em matemática e física e química
df_notas_class = df_notas[
                  (df_notas['df_mat'] != 0) &
                  (df_notas['df_fis'] != 0) &
                  (df_notas['df_qui'] != 0)
]
df_notas_class.head()
````

````python
# Verifica se todos os alunos soram selecionados
len(df_notas_class) + len(df_notas_desclass) == len(df_notas)
````

````python
# Filtra para apenas vacinas administradas na capital do Acre, Rio Branco
df_vacinas_acre = df_vacinas[df_vacinas['paciente_endereco_nmMunicipio'] == "RIO BRANCO"].copy()
df_vacinas_acre.info()
````

### 3.1 Quantas vacinas estão registradas nessa base?

````python
def formata_grande_int(num: int) -> str:
  '''
    Recebe um valor inteiro e, caso ele seja maior que 1.000, formata para
    melhor visualização
    :param num (int): Número que será formatado.
    :return num_formatado (str): Número formatado.
  '''
  num_formatado = f"{num:,.2f}"
  num_formatado = num_formatado.replace(".00", "")
  num_formatado = num_formatado.replace(",", ".")
  return num_formatado
````

````python
# 1) Sua solução aqui
# Usa len() para saber o número de instâncias da base

num_vacinas = formata_grande_int(len(df_vacinas_acre))

print(f"Na base de dados foram registradas {num_vacinas} vacinas aplicadas!")
````

### 3.2 Quantos pacientes foram vacinados? (considere um paciente para cada valor único de paciente_id)

````python
# 2)
'''
Seleciona a coluna paciente_id, que provavelmente contêm dados do paciente que
recbeu a vacina, usa o módulo unique(), para retirar as duplicatas, e o len(),
para saber o número de pacientes
'''
num = formata_grande_int(len(df_vacinas_acre.paciente_id.unique()))

print(f"Foram vacinados {num} pacientes no Acre.")
````

### 3.3 Quantos pacientes únicos tomaram a primeira dose? OBS: Há um caractere especial neste campo. Receba os valores do campo com o método .unique().

````python
# 3)
# Seleciona o valor de primeira dose e usa esse valor para selecionar colunas e valores na DataFrame
seleciona_dose = df_vacinas_acre["vacina_descricao_dose"].value_counts().index[0]
filtra_pacientes = df_vacinas_acre['vacina_descricao_dose'] == seleciona_dose

num = len(df_vacinas_acre[filtra_pacientes]['paciente_id'].unique())
num = formata_grande_int(num)

print(f"{num} pacientes tomaram a primeira dose no Acre.")
````

````python
# 3) solução alternativa
# Seleciona o valor de primeira dose como string e usa esse valor para selecionar colunas e valores na DataFrame

seleciona_dose = '1ª Dose'
filtra_pacientes = df_vacinas_acre['vacina_descricao_dose'] == seleciona_dose

num = len(df_vacinas_acre[filtra_pacientes]['paciente_id'].unique())
num = formata_grande_int(num)

print(f"{num} pacientes tomaram a primeira dose no Acre.")
````

````python
# 3) solução alternativa
# A maioria dos pacientes tom a primeira dose, por isso a função value_counts já retorna o valor desejado

num = df_vacinas_acre[filtra_pacientes]['vacina_descricao_dose'].value_counts()[0]

num = formata_grande_int(num)

print(f"{num} pacientes tomaram a primeira dose no Acre.")
````

### 3.4 Quantos pacientes com menos de 18 anos foram vacinados?

````python
# 4)
# Seleciona os pacientes com menos de 18 anos

filtra_pacientes = df_vacinas_acre['paciente_idade'] < 18

# Filtra da DataFrame esses pacientes e recupera apenas os pacientes, sem duplicatas
num = len(df_vacinas_acre[filtra_pacientes]['paciente_id'].unique())
num = formata_grande_int(num)

print(f"{num} pacientes com menos de 18 anos tomaram vacina no Acre.")
````

### 3.5 Quantos estabelecimentos aplicaram vacina no Acre?

````python
# 5)
# Filtra da DataFrame os estabelecimentos, sem duplicatas

num = len(df_vacinas_acre['estabelecimento_valor'].unique())
num = formata_grande_int(num)

print(f"{num} estabelecimentos aplicaram vacinas no Acre.")
````

## 4) Vacinação II
Gere um *DataFrame* que contenha somente os estabelecimentos que aplicaram vacinas a menores de 18 anos. Nesse *DataFrame* devem conter somente os dados dos estabelecimentos, mais uma coluna sendo a quantidade de vacinas que o estabelecimento aplicou a menores de 18 anos.  
  
1. crie uma cópia do *DataFrame* original, contendo somente os registros de vacinas realizadas a menores de 18 anos.  
2. crie uma lista das colunas desse *DataFrame* com o atributo de *DataFrame* **.columns()**  
3. Nesse *DataFrame* faça uma contagem do campo ```vacina_categoria_nome```.
3. a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.  
4. usando o método *.loc*, selecione somente essas variáveis  
5. Aplique o método **.drop_duplicates** e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento  

````python
# 1)
# Filtra os pacientes com menos de 18 anos
filtra_pacientes = df_vacinas_acre['paciente_idade'] < 18

# Busca os estabelecimentos que administraram vacinas nesses pacientes
df_vacinas_acre_estabelecimento = df_vacinas_acre[filtra_pacientes]['estabelecimento_valor']

# Conta quantas vacinas foram aplicadas por estabelecimentos, instanciando uma DataFrame
df_vacinas_acre_estabelecimento = pd.DataFrame(df_vacinas_acre_estabelecimento.value_counts())

# Obtem uma coluna com os estabelecimentos, sem index
df_vacinas_acre_estabelecimento = df_vacinas_acre_estabelecimento.reset_index()

# Nemomeia as colunas
df_vacinas_acre_estabelecimento.columns = ['estabelecimento_valor','num_doses']
df_vacinas_acre_estabelecimento.head()
````

### 4.1 Crie uma cópia do DataFrame original, contendo somente os registros de vacinas realizadas a menores de 18 anos.

````python
# Instancia uma cópia da DataFrame original, para pacientes com menos de 18 anos
df_vacinas_acre_pacientes = df_vacinas_acre[filtra_pacientes].copy()
df_vacinas_acre_pacientes.info()
````

### 4.2 Crie uma lista das colunas desse DataFrame com o atributo de DataFrame .columns()

````python
# 2)
# Instancia uma lista com os nomes das colunas da DataFrame

lista_colunas = df_vacinas_acre_pacientes.columns
lista_colunas
````

### 4.3 Nesse DataFrame faça uma contagem do campo vacina_categoria_nome.

````python
# 3)
# Conta as vacinas aplicadas por categoria.

df_vacinas_acre_pacientes['vacina_categoria_nome'].value_counts()
````

### 4.4 A partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.

````python
# 4)
# Instancia uma lista com as colunas referêntes aos estabelecimentos

colunas_estabelecimentos = [col for col in lista_colunas if 'estabelecimento_' in col]
colunas_estabelecimentos
````

### 4.5 Usando o método .loc, selecione somente essas variáveis

````python
# 5)
# Seleciona apenas os dados dos estabelecimentos

df_vacinas_acre_pacientes_estabelecimento = df_vacinas_acre_pacientes.loc[:, colunas_estabelecimentos]
df_vacinas_acre_pacientes_estabelecimento.info()
````

### 4.6 Aplique o método .drop_duplicates e crie uma lista com uma linha para cada estabelecimento, com os dados do estabelecimento

````python
# 6)
# Retira as duplicadas dos dados dos estebelecimentos

df_vacinas_acre_pacientes_estabelecimento_drop_duplicates = df_vacinas_acre_pacientes_estabelecimento.drop_duplicates()
df_vacinas_acre_pacientes_estabelecimento_drop_duplicates.info()
````

````python
# Junta os dados de número de vacinas aplicadas por estabelecimentos, com os dados dos estabelecimentos.

df_concat = df_vacinas_acre_estabelecimento.merge(df_vacinas_acre_pacientes_estabelecimento_drop_duplicates)
df_concat.info()
````
